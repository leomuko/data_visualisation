```python
#importing libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#generting even numbers between 0&20
x = range(0,20,2)
x
```

```python
#getting the squares of the even nums in x
y = [i**2 for i in x]
y
```

```python
#barplot
plt.xlabel('even nums')
plt.ylabel('squares')
plt.bar(x,y)
plt.show()
```

```python
#pie plot
plt.pie(y)
plt.legend(y)
```

```python
#scatter plot
plt.scatter(x,y)
```

```python
#line plot
plt.plot(x,y)
plt.show()
```

```python
#Sine and cosine waveform using numpy arrays
x = np.arange(0, 15, 0.1)
fig = plt.subplot()
ax = plt.axes()
ax.plot(x, np.sin(x), label='Sine')
ax.plot(x, np.cos(x), label='Cosine')
ax.legend()
```
